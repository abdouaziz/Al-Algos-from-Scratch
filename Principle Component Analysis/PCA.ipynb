#### Determinant of Matrix

> Determinant of 2x2 Transformation matrix can be intuitively explained as change in area of 2D figure when it is has gone through that transformation. Similar for 3D transformation we can say change in volume after transformation.


> **Example**: Suppose a we have a 2D figure  $ \begin{bmatrix} 2 & 4 & 5 & 3 \\ 2 & 2 & 3 & 3   \end{bmatrix}$ which has an **area of 2**. Also, we have a transformation matrix as  $ \begin{bmatrix} 1 & 2 \\ 1 & 6   \end{bmatrix}$

> After transformation we get $ \begin{bmatrix} 1 & 2  \\ 1 & 6 \end{bmatrix} \begin{bmatrix} 2 & 4 & 5 & 3 \\ 2 & 2 & 3 & 3   \end{bmatrix} =  \begin{bmatrix} 6 & 8 & 11 & 9 \\ 14 & 16 & 23 & 21   \end{bmatrix}$.

> 

>  $ \text{area of transformed figure} = base*height = \sqrt{ (8-6)^2 + (16-14)^2 } * \sqrt{ (13-11)^2+(23-21)^2 } = 2\sqrt2 * 2\sqrt2 = 8 $

> If we calulate determinant of transformation matrix, $ Det \bigg(  \begin{bmatrix} 1 & 2  \\ 1 & 6 \end{bmatrix} \bigg) = 1*6-2*1 = 4 $ 


> **If Determinant of Matrix is 0** : It performs transformation which squeezes space into lower dimenstions or a single point. So area of any region will become zero after transformation.

### Eigen Vectors and Eigen Values

> In many applications, we have a system that takes some input and products an output. A special case of this situation is when the inputs and outputs are vectors and the system of linear transformation(reoresented by A)

> So, if the input vector is '$x$', then output is '$Ax$'. **Usually the direction of output $Ax$ is different from the direction of input $x$**. To understand system better an important question is to ask: what are the input vectors which do not change the direction when they pass through the system? It is okay if magnitude changes but the direction shouldn't. **In other words, what are x's for which Ax is just a scaler multiple of x**?. These x's are precisely eigenvectors.

> **Eigen Vector**: If the direction of a input vector '$x$' does not change after the going through trasformation matrix '$A$', then $x$ is a eigen vector. So, $Ax$ will be just a scaler multiple of '$x$'. Intuitively, **Eigen Vector is axis of rotation of a transformation.**

> **Eigen Value** : Each eigen vector is associated with a eigen value which a factor by which eigen vector will be streched or squeeshed during transformation. In other words, it is the scaler multiple of eigen vector after transformation '$A$'. $ Ax = \lambda x $, where 'x' is a eigen vector and '$\lambda$' is its eignen value.

$$ A \overrightarrow v = \lambda \overrightarrow v $$

$$ A \overrightarrow v = (\lambda I) \overrightarrow v $$

$$ (A - \lambda I) \overrightarrow v = \overrightarrow 0 $$

It is only possible for a non-zero vector to become zero if the transforamtion $(A - \lambda I)$ squeezes space in lower dimensions, i.e $Det(A - \lambda I) = 0$

$$ Det\big( (A - \lambda I) \big) = 0 $$

Calculate possible values of $\lambda$, and for each value of $\lambda$ find out  $ \overrightarrow v $ using below equation.

$$ (A - \lambda I) \overrightarrow v = \overrightarrow 0 $$

> **Example**: For $ A =  \begin{bmatrix} 3 & 1 \\ 0 & 2 \end{bmatrix}$, $ Det( \begin{bmatrix} 3-\lambda & 1 \\ 0 & 2-\lambda \end{bmatrix} ) = 0 $

> $ (3-\lambda)(2-\lambda) - 1*0 = 0,\;\;\;\; \lambda = 3,2 $

> Now, $(A - \lambda I) \overrightarrow v = \overrightarrow 0$,  $ \begin{bmatrix} 3-\lambda & 1 \\ 0 & 2-\lambda \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} $

> $ \begin{bmatrix} (3-\lambda)x+y \\ (2-\lambda)y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} $

> $ \text{ For, $\lambda$ = 3 }, y = 0 $

> $ \text{ For, $\lambda$ = 2 }, y = -x $

> **Properties**
> 1. It is not necessory for a 2D transformation to have atleast one eigen vectors, say rotation by 90 degrees ($\begin{bmatrix} 0 & -1 \\ -1 & 0 \end{bmatrix}$)  which has no eigen vectors
> 2. It is possible to have multiple eigen vectors but a single eigen value. For example  ($\begin{bmatrix} 2 & 0 \\ 0 & 2 \end{bmatrix}$), here only single eigen value 2 is there. But every vector in plane is eigen vector
> 3. A square matrix A and its transpose have the same eigenvalues.
> 4. A square matrix A of order n will not have more than n eigenvalues but can have infinite number of eigenvectors.

### Dimensionality Reduction using PCA

> **PCA (Principle Component Analysis)** : PCA basically is a projection of a higher-dimensional space into a lower dimensional space while keeping the total variance as high as possible.

> PCA constructs some new characteristics that turn out to summarize our data. Of course these new characteristics are constructed using the old ones; (we call them linear combinations). Also PCA looks for properties that allow to reconstruct the original characteristics as well.

> Mathematically this can be done by calculating the so-called eigenvectors of the covariance matrix, after that we use just use the n eigenvectors with the biggest eigenvalues to project the object into n-dimensional space

$$ \big[ \textit{Covariance Matrix} \big].\big[ \textit{Eigen Vectors} \big] = \big[ \textit{Eigen Values} \big].\big[\textit{Eigen Vectors}\big] $$

> [Nice Article](https://towardsdatascience.com/a-one-stop-shop-for-principal-component-analysis-5582fb7e0a9c) [Nice Post](https://stats.stackexchange.com/questions/2691/making-sense-of-principal-component-analysis-eigenvectors-eigenvalues?answertab=votes#tab-top)

> **LDA vs PCA** : PCA just ignores the class labels while **LDA (Linear Discriminant Analysis)** aims to find the directions that maximize the separation (or discrimination) between different classes

> Most of the implementations of PCA uses **SVD (Singular Value Decomposition)**. The singular value decomposition (SVD) provides another way to factorize a matrix, into singular vectors and singular values. The SVD allows us to discover some of the same kind of information as the eigendecomposition. However, the SVD is more generally applicable.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def covariance(X): 
    x = X - X.mean(axis=0)
    return np.dot(x.T,x)/len(x)
```

```python
from sklearn.datasets import load_digits
data =  load_digits()
X,y = data.data, data.target
X.shape, y.shape
```

```python
cov_mat = covariance(X)
cov_mat.shape
```

```python
values,vectors = np.linalg.eig(cov_mat)
```

```python
k=2
args = np.argsort(values)[-k:]
args
```

```python
X_pca = np.dot( X, vectors[args].T )
X_pca.shape
```
